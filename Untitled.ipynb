```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
t = np.linspace(-0.02,0.05,1000)
x = 325*np.sin(2*np.pi*50*t)
plt.plot(t,x)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title(r'Plot CT signal $x(t)=325 \sin(2\pi 50 t)$')
plt.xlim([-0.02,0.05])
```

```python
n = np.arange(50)
dt = 0.07/50
x = np.sin(2 * np.pi * 50 * n *dt)
plt.xlabel('n')
plt.ylabel('x[n]')
plt.title(r'Plot DT signal $x[n] = 325 \sin(2\pi 50 n \Delta t)$')
plt.stem(n,x)
```
